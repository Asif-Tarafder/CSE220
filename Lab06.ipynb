```python
#task-6
#binarry search

def binary_search(a,val,l,r):
    if l>r:
        return("not found")
    elif l<=r:
        m=(l+r)//2
        if a[m]==val:
            return m
        elif val>a[m]:
            l=m+1
        else:
            r=m-1
        m=binary_search(a,val,l,r)
        if m!=None:
            return(m)
        
a=[-9, 0, 1, 2, 3, 7, 9, 13, 17, 19, 55, 56, 67]   # sorted list
n=int(input())                                     # which num you want to find
print(binary_search(a,n,0,len(a)-1))
```
