```python
#Task-02_
class Node:
    def __init__(self,element,nxt,prev):
        self.element=element
        self.next=nxt
        self.prev=prev

# 1,2 স্যার দেখিয়ে দিয়েছিলেন ল্যাবে। তাই যদি অন্য কারো সাথে মিলে যায় তাহলে আমি দায়ী নই। দয়া করে, নিজগুণে মার্ক্স দিয়ে দিবেন #
class MyList:
    
##################################################################

           #task-2(01)
    
    def __init__(self,array):
        
        # dummy node
        self.head=Node(None,None,None) 
        
        # dummy headed circular list
        tail=self.head
        tail.next=self.head
        self.head.prev=tail
        
        # is list empty or not?
        if len(array)==0:
            print("Array is empty try again")
            
        # dummy headed list
        else:
            for i in array:
                node=Node(i,None,None)
                tail.next=node
                node.prev=tail
                tail=node
            #creat circular
            tail.next=self.head
            self.head.prev=tail
            
##########################################################################

                #task-2(02)
    
    def showList(self):
        if (self.head.prev==self.head) and (self.head.next==self.head):
            print("Empty List")
        else:
            tail=self.head.next
            while tail!=self.head:
                if tail.next==self.head:
                    print(tail.element)
                else:
                    print(tail.element,end="<--->")
                tail=tail.next

########################################################################

            #task-2(03)

    def insert_01(self, newElement):
        tail=self.head.next
        
        # for check duplicate 
        while tail!=self.head:
            if tail.element==newElement:
                return(" the key already exists ")
            tail=tail.next
            
        # for insert in the last 
        tail=self.head.prev
        node=Node(newElement,None,None)
        tail.next=node
        node.prev=tail
        node.next=self.head
        self.head.prev=node
        
##########################################################################  

            #task-2(04)
    
    def insert_02(self, newElement,index):
        # Is list empty or not?
        if (self.head.prev==self.head) and (self.head.next==self.head):
            return("Empty List try again")
        
        # index count and is element already exsist?
        tail=self.head.next
        count=0
        while tail!=self.head:
            if tail.element==newElement:
                return(" the key already exists ")
            tail=tail.next
            count=count+1
        if (index<0 or index>count):
            return("index out of bound")
        
        #insert
        else:
            tail=tail.next
            for i in range(count+1):
                if i==index:
                    node=Node(newElement,None,None)
                    tail.prev.next=node
                    node.prev=tail.prev
                    node.next=tail
                    tail.prev=node
                tail=tail.next
                
###################################################################

             #task-2(05)

    def remove(self, index):
        # Is list empty or not?
        if (self.head.prev==self.head) and (self.head.next==self.head):
            return("Empty List try again")
        
        # index count and is it exsist the limit or not?
        tail=self.head.next
        count=0
        while tail!=self.head:
            tail=tail.next
            count=count+1
        if (index<0 or index>=count):
            return("index out of bound")
        
        #remove
        else:
            tail=tail.next
            for i in range(count+1):
                if i==index:
                    tail.prev.next=tail.next
                    tail.next.prev=tail.prev
                    break
                tail=tail.next
                
################################################################### 

               #task-2(06)
    
    def removeKey(self, deletekey):
        tail=self.head.next
        b=0
        while tail!=self.head:
            if tail.element==deletekey:
                tail.prev.next=tail.next
                tail.next.prev=tail.prev
                b=1
                return(tail.element)
            tail=tail.next
        if b==0:
            return("deletekey is not present in list")
        
#################################################################### 
   
         #taster
print("My Dummy Double linked list")        
array=[11,22,33,44,55,66]
obj=MyList(array)
obj.showList()
print()
###############################

  #insertion at last index
    
print("Insertion at last index")
re=obj.insert_01(77)
if re!=None:
    print(re)
else:
    obj.showList() 
print()   
################################
    
    #insertion at any index
    
print("Insertion at any index:")
re=obj.insert_02(88,7)
if re!=None:
    print(re)
else:
    obj.showList()   
print()

#################################

   #remove from any index
    
print("remove from any index")
re=obj.remove(5)
if re!=None:
    print(re)
else:
    obj.showList() 
print()

##################################

   #remove any deletekey
print("remove any deletekey:")    
print(obj.removeKey(33))

##################################

```
