```python
class Node:
    def __init__(self,element,parents=None,left=None,right=None):
        self.element=element
        self.parent=parents
        self.left=left
        self.right=right
           
    
class Trees:
    def __init__(self):
        self.root=None
    
    def Tree_Maker(self,bin_arr,idx=1): 
        
        if idx>len(bin_arr)-1 or bin_arr[idx]==None:
            return None
        
        else:
            root_node=Node(bin_arr[idx])
            root_node.left=self.Tree_Maker(bin_arr,idx*2)
            root_node.right=self.Tree_Maker(bin_arr,idx*2+1)
            
            if root_node.left!=None:
                root_node.left.parent=root_node
                
            if root_node.right!=None:
                root_node.right.parent=root_node
                
        return(root_node)  # root return korbay
    

##################################################################



               #Task-1

    def Height_0f_tree(self,root_node):
        if root_node==None:
            return 0
        else:
            return 1+self.Found_max(self.Height_0f_tree(root_node.left),self.Height_0f_tree(root_node.right))
        
    def Found_max(self,left,right):
        
        if left>right:
            return left
        else:
            return right
        
        
        
######################################################################### 

              #Task-2

    def Level_count(self,node):
        if node==None:
            return("level can not find as this node is not exist in the tree. Please, try again later")
            
        elif node.parent==None:
            return 0
        
        else:
            return 1+self.Level_count(node.parent)

#########################################################################

                 #Task-03

    def Pre_order(self,root_node):
        
        if root_node!=None:            
            print(root_node.element,end=" ")
            self.Pre_order(root_node.left)
            self.Pre_order(root_node.right)
            
        #print()  
#######################################################################   

                  #Task-4
    
    def In_order(self,root_node):
        
        if root_node!=None:            
            self.In_order(root_node.left)
            print(root_node.element,end=" ")
            self.In_order(root_node.right)
        #print()  
            
########################################################################



                  #Task-5
    
    def Post_order(self,root_node):
        
        if root_node!=None:            
            self.Post_order(root_node.left)
            self.Post_order(root_node.right)
            print(root_node.element,end=" ")
        #print()  

###########################################################################


                  #Task-6
    
    def checker(self,node1,node2):
        if self.Height_0f_tree(node1)!=self.Height_0f_tree(node2):
            return("Two tree are not same")
        else:
            #print(self.copy_check(node1,node2))
            return(self.copy_check(node1,node2))
    
    def copy_check(self,node1,node2):       
        if (node1!=None and node2!=None):
            if (node1.element==node2.element):
                #print(node1.element,node2.element)
                left=(self.copy_check(node1.left,node2.left))
                #print(left)
                if left=="True":
                    return("Two tree are not same")
            
                right=(self.copy_check(node1.right,node2.right))
                #print(right)
                if right=="True":
                    return("Two tree are not same")
                
            #print(node1.element,node2.element)
            elif (node1.element!=node2.element):
                #print(node1.element,node2.element)
                return("True")
        

###########################################################################

           #task-7
    
    def Tree_coping(self,root_node):
        
        if (root_node == None):
            return
        
        new_node=Node(root_node.element,None,None,None)
        self.tree_coping(root_node,new_node)
        return(new_node)

    def tree_coping(self, root_node,new_node):
        if root_node.left!=None:
            new_node.left=Node(root_node.left.element,None,None,None)
            new_node.left.parent = new_node
            self.tree_coping(root_node.left,new_node.left)
            
        if root_node.right!=None:
            new_node.right = Node(root_node.right.element,None,None,None)
            new_node.right.parent = new_node
            self.tree_coping(root_node.right,new_node.right)
    
##########################################################################        

#binary_array=[None,"A","B","C","D","E","F","G",None,None,None,None,None,"H",None,None]
binary_array=[None,"P","T","X",None,"S","Q","U",None,None,"Y","W","R",None,None,"V"]
obj=Trees()
root=obj.Tree_Maker(binary_array)
print("Height of the tree:",obj.Height_0f_tree(root))
print("The node is in level:",obj.Level_count(root.right.left.right))
print("Preorder :",end=" ")
obj.Pre_order(root)
print("\nIn_order :",end=" ")
obj.In_order(root)
print("\nPostorder:",end=" ")
obj.Post_order(root)

#---------------------------------------------------------------------------------------------#

           # Two tree are same or not(task-6)
    
#binary_array1=[None,"A","B","C","D","E","F","G",None,None,None,None,None,"H",None,None]
#binary_array1=[None,"A","B","C","D","E"]
#binary_array1=[None,"A","B","K","D","E"]
binary_array2=[None,"P","T","X",None,"S","Q","U",None,None,"Y","W","R",None,None,"V"]
binary_array2=[None,"A","B","K","D","E"]
obj1=Trees()
obj2=Trees()
root1=obj1.Tree_Maker(binary_array1)
root2=obj2.Tree_Maker(binary_array2)
print()
ans=obj1.checker(root1,root2)
if ans==None:
    print("Two tree are excetly same")
else:
    print(ans)
#--------------------------------------------------------------------------------------------------#

print()
obj3=Trees()
root3=obj.Tree_coping(root)
#print(root3.element)
print("Height of the copy tree:",obj3.Height_0f_tree(root3))
print("The node of copy tree is in level:",obj.Level_count(root3.right.left.right))
print("Preorder of copy tree :",end=" ")
obj3.Pre_order(root3)
print("\nIn_order of copy tree :",end=" ")
obj3.In_order(root3)
print("\nPostorder of copy tree:",end=" ")
obj3.Post_order(root3)
```
