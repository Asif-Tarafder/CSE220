```python
#Task - 01
class Stack_Array:
    
    #my constructor
    def __init__(self,s):
        self.stack=[]
        start=["(","{","["]
        end=[")","}","]"]
        boolien=0
        
        #to find out error index
        self.possition=[]
        
        #to find opening bracket and endding bracket
        for i in range(len(s)):
            
            if s[i] in start:
                self.push(s[i])
                self.possition.append(i+1)
                
                
            elif s[i] in end:
                b=self.pop()
                if b==1:
                    print(s,"\nThis expression is NOT correct.","\nError at character #",i+1,".'",s[i],"' not opened.")
                    boolien=1
                    break
                elif chr(b+1)==s[i] or chr(b+2)==s[i] :
                    b=0
                else:
                    print(s,"\nThis expression is NOT correct.","\nError at character #",self.possition[-1],".'",chr(b),"' not closed.")
                    boolien=1
                    break
                self.possition=self.possition[0:-1]
                    
        # To check after work, is stack empty or not? 
        if len(self.stack)==0 and boolien==0:
            print(s,"\nThis expression is correct.")
        elif boolien!=1:
            print(s,"\nThis expression is not correct. \nError at character #",self.possition[-1]+1,".'",self.stack[-1],"' not closed.")
    
    
    
    def push(self,bracket):
        self.stack.append(bracket)

        
    def pop(self):        
        if (len(self.stack))==0:
            return 1
        else:
            r=self.stack[-1]
            self.stack=self.stack[0:-1]
            return(ord(r))
            
n=input()
print("\nOutput:")
obj=Stack_Array(n)
            
            
```
